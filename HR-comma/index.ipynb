# 4) Data Standardisation

```python
# In [21]: Data Standardisation

from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(X)
```

# 5) Computing Eigenvectors and Eigenvalues:


```python
# In [22]: Computing Eigenvectors and Eigenvalues
# Covariance matrix
mean_vec = np.mean(X_std, axis=0)
cov_mat = (X_std - mean_vec).T.dot((X_std - mean_vec)) / (X_std.shape[0]-1)
print('Covariance matrix \n%s' %cov_mat)
```

```python
# In [23]:
print('NumPy covariance matrix: \n%s' %np.cov(X_std.T))
```

```python
# In [24]: Heatmap 
plt.figure(figsize=(8,8))
sns.heatmap(cov_mat, vmax=1, square=True,annot=True,cmap='cubehelix')

plt.title('Covariance between different features')
```

```python
# In [25]
eig_vals, eig_vecs = np.linalg.eig(cov_mat)

print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

```python
# In [26]

# Make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eig_pairs.sort(key=lambda x: x[0], reverse=True)

# Visually confirm that the list is correctly sorted by decreasing eigenvalues
print('Eigenvalues in descending order:\n')
for i in eig_pairs:
    print(i[0])
```

```python
# In [27]

tot = sum(eig_vals)
var_exp = [(i / tot)*100 for i in sorted(eig_vals, reverse=True)]
var_exp
```

```python
# In [28]

with plt.style.context('dark_background'):
    plt.figure(figsize=(6, 4))
    plt.bar(range(7), var_exp, alpha=0.5, align='center', label='Explained variance')
    plt.ylabel('Variance ratio')
    plt.xlabel('n_componets (principal components)')
```

```python
# In [29]

matrix_w = np.hstack((eig_pairs[0][1].reshape(7,1), 
                      eig_pairs[1][1].reshape(7,1)
                    ))
print('Matrix W:\n', matrix_w)
```

```python
# In [30]

Y = X_std.dot(matrix_w)
Y
```

```python
# In [31]

# PCA in scikit-learn
from sklearn.decomposition import PCA
pca = PCA().fit(X_std)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlim(0,7,1)
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
```
